```python
#inserimento DB

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data_path = "C://Users//Account Lavoro//Desktop//EPICODEDB//owid-covid-data.csv"
df = pd.read_csv(data_path,low_memory=False)
df

```

```python
#Domanda1.A/1.B

#dimensione dataset
df.shape

#metadati
df.columns


```

```python
#Domanda 2
casi_continente = df.groupby('continent')['total_cases'].max().reset_index()
casi_continente = casi_continente[casi_continente['continent'].notna()]
print(casi_continente)
```

```python
#Domanda 3
continenti = ['Europe', 'Asia']
dati_continenti = df[df['continent'].isin(continenti)]

# Calcolo il numero totale di casi per EU/AS
casi_continente = dati_continenti.groupby('continent')['new_cases'].max().reset_index()
print(casi_continente)

# Calcolo media
casi_continente_media = dati_continenti.groupby('continent')['new_cases'].mean().reset_index()
print(casi_continente_media)
```

```python
# Calcolo la percentuale dei casi totali per Europa e Asia

dati_continente = df.dropna(subset=['continent'])

# dati per il continente Europa
europe_data = df[df['continent'] == 'Europe']

total_new_cases_europe = europe_data['new_cases'].sum()
total_new_cases_world = dati_continente['new_cases'].sum()

incidenza_eu = (total_new_cases_europe/total_new_cases_world)*100
print("Sui casi totali mondiali", total_new_cases_world, "l'europa incide per il", incidenza_eu,"con un totale di ", total_new_cases_europe)

#---------------------------------------------------------------------------------------------------------------------

# dati per il continente Asia
asia_data = df[df['continent'] == 'Asia']

total_new_cases_asia = asia_data['new_cases'].sum()
total_new_cases_world = dati_continente['new_cases'].sum()

incidenza_as = (total_new_cases_asia/total_new_cases_world)*100
print("Sui casi totali mondiali", total_new_cases_world, "l'asia incide per il", incidenza_as,"con un totale di ", total_new_cases_asia)
```

```python
#Domanda4
# Italia 22
italia_2022 = df[(df['location'] == 'Italy') & (df['date'].str.contains('2022'))]
italia_2022['date'] = pd.to_datetime(italia_2022['date'])
italia_2022.set_index('date', inplace=True)
italia_2022_mensile = italia_2022.resample('M').last().reset_index()

# Grafico dell'evoluzione dei casi totali in Italia per il 2022
plt.figure(figsize=(14, 7))
sns.lineplot(data=italia_2022_mensile, x='date', y='total_cases')
plt.xticks(rotation=45)
plt.title('Casi Totali in Italia (2022)')
plt.xlabel('Data')
plt.ylabel('Casi Totali')
plt.show()

# Grafico dell'evoluzione dei nuovi casi in Italia per il 2022
italia_2022_mensile['new_cases_mensili'] = italia_2022['new_cases'].resample('M').sum().values
plt.figure(figsize=(14, 7))
sns.lineplot(data=italia_2022_mensile, x='date', y='new_cases_mensili')
plt.xticks(rotation=45)
plt.title('Nuovi Casi in Italia (2022)')
plt.xlabel('Data')
plt.ylabel('Nuovi Casi')
plt.show()

# Somma cumulativa dei nuovi casi per il 2022
italia_2022['cumulative_new_cases'] = italia_2022['new_cases'].cumsum()
italia_2022_mensile['cumulative_new_cases'] = italia_2022['cumulative_new_cases'].resample('M').last().values
plt.figure(figsize=(14, 7))
sns.lineplot(data=italia_2022_mensile, x='date', y='cumulative_new_cases')
plt.xticks(rotation=45)
plt.title('Casi Nuovi Cumulativi in Italia (2022)')
plt.xlabel('Data')
plt.ylabel('Nuovi Casi Cumulativi')
plt.show()
```

```python
#Esercizio6
# Filtrare i dati per le nazioni specificate nel 2023
nazioni = ['Italy', 'Germany', 'France', 'Spain']
ospedalizzazione_dati = df[(df['location'].isin(nazioni)) & 
                           (df['date'] >= '2023-01-01') & 
                           (df['date'] <= '2023-12-31')]

# Convertire la colonna delle date in formato datetime
ospedalizzazione_dati['date'] = pd.to_datetime(ospedalizzazione_dati['date'])

# Filtrare per mesi
ospedalizzazione_dati.set_index('date', inplace=True)
ospedalizzazione_dati_mensile = ospedalizzazione_dati.resample('M').sum().reset_index()

# Sommare il numero di ospedalizzazioni per nazione nel 2023
ospedalizzazione_somma = ospedalizzazione_dati_mensile.groupby('location')['hosp_patients'].sum().reset_index()

# Grafico delle ospedalizzazioni per nazione nel 2023
plt.figure(figsize=(14, 7))
sns.barplot(data=ospedalizzazione_somma, x='location', y='hosp_patients')
plt.title('Ospedalizzazioni Totali nel 2023 per Nazione')
plt.xlabel('Nazione')
plt.ylabel('Numero di Pazienti Ospedalizzati')
plt.show()
```
